### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
# Take input from user and convert to lakhs
ctc_lakhs = float(input('Enter ctc in lakhs : '))
ctc = ctc_lakhs * 1000000

# Salary taxes
hra = ctc * 0.10
da = ctc * 0.05
pf = ctc * 0.03

# Total salary after deductions
salary_after_deductions = ctc - (hra + da + pf)

if ctc_lakhs < 5:
    tax = 0
elif ctc_lakhs < 10:
    tax = salary_after_deductions * 0.10
elif ctc_lakhs < 20:
    tax = salary_after_deductions * 0.20
else:
    tax = salary_after_deductions * 0.30

# Annual Salary
annual_inhand = salary_after_deductions - tax
monthly_salary = annual_inhand / 12

print("Monthly In-hand Salary:", monthly_salary)

````

### `Problem 5` - Exercise 12: Display Fibonacci series up to 10 terms.

*Note: The Fibonacci Sequence is a series of numbers. The next number is found by adding up the two numbers before it. The first two numbers are 0 and 1. For example, 0, 1, 1, 2, 3, 5, 8, 13, 21. The next number in this series above is 13+21 = 34*

````python
# Number of terms
n = 10

# First two terms
a = 0
b = 1

print("Fibonacci Series:")

# Print first two terms
print(a, b, end=" ")

# Generate remaining terms
for i in range(2, n):
    c = a + b
    print(c, end=" ")
    a = b
    b = c
````

### `Problem 6` - Find the factorial of a given number.

Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: `!`) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```bash
5! = 5 × 4 × 3 × 2 × 1 = 120
```

Output:

```bash
120
```

````python
# Take input from user
n = int(input("Enter a number: "))

factorial = 1

for i in range(1, n + 1):
    factorial = factorial * i

print("Factorial:", factorial)
````

```bash
76542
```

`Output:`

```bash
24567
```

````python
# Take input from user
n = int(input("Enter a number: "))

reverse = 0

while n > 0:
    digit = n % 10
    reverse = reverse * 10 + digit
    n = n // 10

print("Reversed number:", reverse)
````

### `Problem 8`: Take a user input as integer N. Find out the sum from 1 to N. If any number if divisible by 5, then skip that number. And if the sum is greater than 300, don't need to calculate the sum further more. Print the final result. And don't use for loop to solve this problem.

**Example 1:**

`Input:`

```bash
30
```

`Output:`

```bash
276
```

````python
# Take input
N = int(input("Enter a number: "))

sum_total = 0
i = 1

while i <= N and sum_total <= 300:
    if i % 5 != 0:
        sum_total += i
    i += 1

print("Final sum:", sum_total)
````

### `Problem 9`: Write a program that keeps on accepting a number from the user until the user enters Zero. Display the sum and average of all the numbers.

````python
# Initialize sum and count
total = 0
count = 0

while True:
    num = float(input("Enter a number (0 to stop): "))

    if num == 0:
        break

    total += num
    count += 1

# Calculate average
if count > 0:
    average = total / count
else:
    average = 0

print("Sum of numbers:", total)
print("Average of numbers:", round(average, 2))
````

###`Problem 9`: Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.

````python
# Initialize empty list to store results
result = []

# Loop from 2000 to 3200 inclusive
for num in range(2000, 3201):
    if num % 7 == 0 and num % 5 != 0:
        result.append(str(num))  # Convert number to string for joining

# Print as comma-separated sequence
print(",".join(result))
````

###`Problem 10`: Write a program, which will find all such numbers between 1000 and 3000 (both included) such that each digit of the number is an even number. The numbers obtained should be printed in a space-separated sequence on a single line.



````python
# Initialize empty list to store numbers
result = []

# Loop from 1000 to 3000 inclusive
for num in range(1000, 3001):
    num_str = str(num)  # Convert number to string
    all_even = True     # Flag to check if all digits are even

    for digit in num_str:
        if int(digit) % 2 != 0:  # If any digit is odd
            all_even = False
            break                # No need to check further

    if all_even:
        result.append(num_str)

# Print all numbers space-separated
print(" ".join(result))
````

###`Problem 11`: A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps.
The trace of robot movement is shown as the following:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
> The numbers after the direction are steps.

> `!` means robot stop there.

**Please write a program to compute the distance from current position after a sequence of movement and original point.**

*If the distance is a float, then just print the nearest integer.*

Example:

`Input`:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
`Output`:
```
2
```

````python
import math

x = 0
y = 0

while True:
    s = input("Enter direction and steps (or '!' to stop): ")

    if s == "!":
        break

    direction, steps = s.split()
    steps = int(steps)

    if direction.upper() == "UP":
        y += steps
    elif direction.upper() == "DOWN":
        y -= steps
    elif direction.upper() == "LEFT":
        x -= steps
    elif direction.upper() == "RIGHT":
        x += steps

# Compute distance from origin
distance = math.sqrt(x**2 + y**2)

# Print nearest integer
print(round(distance))
````

###`Problem 12`:Write a program to print whether a given number is a prime number or not

````python
import math

# Take input
n = int(input("Enter a number: "))

if n <= 1:
    print(n, "is not a prime number")
else:
    is_prime = True
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            is_prime = False
            break
    if is_prime:
        print(n, "is a prime number")
    else:
        print(n, "is not a prime number")
````

###`Problem 13`:Print all the Armstrong numbers in a given range.
Range will be provided by the user<br>
Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

````python
# Take range input
start = int(input("Enter start of range: "))
end = int(input("Enter end of range: "))

print("Armstrong numbers in the range:")

for num in range(start, end + 1):
    sum_of_cubes = 0
    temp = num  # Store original number

    while temp > 0:
        digit = temp % 10
        sum_of_cubes += digit ** 3
        temp //= 10

    if sum_of_cubes == num:
        print(num, end=" ")
````

###`Problem 14`:Calculate the angle between the hour hand and minute hand.

Note: There can be two angles between hands; we need to print a minimum of two. Also, we need to print the floor of the final result angle. For example, if the final angle is 10.61, we need to print 10.

Input:<br>
H = 9 , M = 0<br>
Output:<br>
90<br>
Explanation:<br>
The minimum angle between hour and minute
hand when the time is 9 is 90 degress.

````python
# Input hours and minutes
H = int(input("Enter hours (0-12): "))
M = int(input("Enter minutes (0-59): "))

# Calculate angles
minute_angle = M * 6
hour_angle = H * 30 + M * 0.5

# Angle between hands
angle = abs(hour_angle - minute_angle)

# Minimum angle
min_angle = min(angle, 360 - angle)

# Print floor of result
print(int(min_angle))
````
