# **Literals**

```python
#Inside the variable the raw value called literals
#A value given to variable are literals.

#Integers
a = 0b1010 #Binary literals
b = 100 #Decimal literals
c = 0o310 #Octal literals
d = 0x12c #Hexadecimal literals

#Float literals
float_1 = 10.5
float_2 = 1.5e2 #1.5 * 10^2
float_3 = 1.5e-3 #1.5 * 10^-3

#Complex literals
x = 3.14j

print(a,b,c,d)
print(float_1,float_2,float_3)
print(x,x.imag,x.real)
```

```python
string = 'This is python'
strings = "This is python"
char = "C"
multiline_str = """This is multi line string with more than 1 line code"""
unicode = u"\U0001f600\U0001F606\U0001F923"
raw_str = r"raw \n string"
print(string)
print(strings)
print(char)
print(unicode)
print(raw_str)
```
