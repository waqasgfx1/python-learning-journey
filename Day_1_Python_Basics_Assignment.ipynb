# **Program to convert Celcius to Fahrenheit**

```
#Celcius to Fahrenheit
celcius = float(input("Enter the temperature in Celcius: "))
fahrenheit = (celcius * 9/5) + 32
print("Temperature in Fahrenheit: ",fahrenheit)
```

# **Take 2 numbers as input from the user.Write a program to swap the numbers without using any special python syntax.**

```
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
temp = a
a = b
b = temp
print("After swapping:",end = '\n')
print("First number: ",a)
print("Second number: ",b)
```

# **Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.**

```
#Taking input from user
x1 = float(input("Enter the x1 coordinate: "))
x2 = float(input("Enter the x2 coordinate: "))

y1 = float(input("Enter the y1 coordinate: "))
y2 = float(input("Enter the y2 coordinate: "))

#Calculate Euclidean distance
distance = ((x2-x1)**2 + (y2-y1)**2) / 0.5

#Displaying the result
print("Euclidean distance between two coordinates: ",distance)
```

# **Write a program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.**

```
#Taking input from the user
principle = float(input("Enter the principle amount:"))
rate = float(input("Enter the rate of interest:"))
time = float(input("Enter the time period: "))

#Calculate simple interest
simple_interest = (principle * rate * time) / 100

#Displaying the result
print("Simple interest: ",simple_interest)
```

# **Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.**
For example: Input: heads -> 4 legs -> 12
Output: dogs -> 2 chicken -> 2**

```
# Taking input from the user
heads = int(input("Enter total number of heads: "))
legs = int(input("Enter total number of legs: "))

# Calculating number of dogs and chickens
dogs = (legs - 2 * heads) // 2
chickens = heads - dogs

# Displaying the result
print("Dogs:", dogs)
print("Chickens:", chickens)
```

#  Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

```
# Taking input from the user
n = int(input("Enter the value of n: "))

sum_of_squares = 0

# Calculating sum of squares
for i in range(1, n + 1):
    sum_of_squares += i ** 2

# Displaying the result
print("Sum of squares of first", n, "natural numbers is:", sum_of_squares)
```

#  Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

```
# Taking input from the user
a = float(input("Enter the first term: "))
b = float(input("Enter the second term: "))
n = int(input("Enter the value of n: "))

# Calculating common difference
d = b - a

# Calculating nth term
nth_term = a + (n - 1) * d

# Displaying the result
print("The", n, "th term of the arithmetic series is:", nth_term)
```

# Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user

```
# Taking input from the user
a = int(input("Enter numerator of first fraction: "))
b = int(input("Enter denominator of first fraction: "))

c = int(input("Enter numerator of second fraction: "))
d = int(input("Enter denominator of second fraction: "))

# Calculating sum of fractions
numerator = (a * d) + (c * b)
denominator = b * d

# Displaying the result
print("Sum of the two fractions is:", numerator, "/", denominator)
```

#  Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.
Input:
Dimensions of the milk tank
H = 20cm, L = 20cm, B = 20cm

Dimensions of the glass
h = 3cm, r = 1cm

```
import math

# Input: dimensions of the milk tank
H = float(input("Enter height of tank (cm): "))
L = float(input("Enter length of tank (cm): "))
B = float(input("Enter breadth of tank (cm): "))

# Input: dimensions of the glass
h = float(input("Enter height of glass (cm): "))
r = float(input("Enter radius of glass (cm): "))

# Volume calculations
tank_volume = H * L * B
glass_volume = math.pi * r * r * h

# Number of glasses
number_of_glasses = tank_volume // glass_volume

# Output
print("Number of glasses of milk that can be obtained:", int(number_of_glasses))
```
